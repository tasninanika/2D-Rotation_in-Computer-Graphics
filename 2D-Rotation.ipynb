```python
import matplotlib.pyplot as plt
import math

# Step 1: Original square points
square = [(1, 1), (1, 4), (4, 4), (4, 1)]

# Step 2: Rotation angle (in degrees)
angle_deg = 45
angle_rad = math.radians(angle_deg)  # convert to radians

# Step 3: Rotate the square
rotated_square = []
for x, y in square:
    new_x = x * math.cos(angle_rad) - y * math.sin(angle_rad)
    new_y = x * math.sin(angle_rad) + y * math.cos(angle_rad)
    rotated_square.append((new_x, new_y))

# Step 4: Close the shapes
square.append(square[0])
rotated_square.append(rotated_square[0])

# Step 5: Separate x and y values
x1, y1 = zip(*square)
x2, y2 = zip(*rotated_square)

# Step 6: Plot both
plt.plot(x1, y1, 'go-', label='Original Square')
plt.plot(x2, y2, 'o-', label=f'Rotated Square ({angle_deg}Â°)')

# Extras
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Rotation of a Square")
plt.legend()
plt.axis('equal')
plt.show()
```
